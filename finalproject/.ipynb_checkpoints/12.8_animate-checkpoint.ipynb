```python
import numpy as np
import matplotlib.pyplot as plt
import sys
import os
sys.path.append(os.path.join('..', 'importantfunctions'))
from ODEsolvers import *
import matplotlib.animation as animation
```

```python
#this is the diffeq that does the affect of body b on a
def Fab(ra,rb,t,mb,G=1): #where ra is the vector (x,y,vx,vy) for body a and rb is for body b
    dx=rb[0]-ra[0]
    dy=rb[1]-ra[1]
    
    vx=ra[2]
    vy=ra[3]
    
    R=(dx**2+dy**2)**0.5

    ax=G*mb*dx/(R**3)
    ay=G*mb*dy/(R**3)
    
    return np.array([vx,vy,ax,ay],float)
```

```python
def diffeq(state,t,m1=1,m2=1,m3=1):
    r1=state[0:4]
    r2=state[4:8]
    r3=state[8:12]

    dr1=Fab(r1,r2,t,m2)+Fab(r1,r3,t,m3) #just adding up the effects of the different stars on each other
    dr2=Fab(r2,r1,t,m1)+Fab(r2,r3,t,m3)
    dr3=Fab(r3,r1,t,m1)+Fab(r3,r2,t,m2)

    return np.concatenate([dr1,dr2,dr3])
```

```python
#this is the butterfly I sequence orbit 1.2.A
R01=np.array([-1,0,0.306893,0.125507])
R02=np.array([1,0,0.306893,0.125507])
R03=np.array([0,0.3133550361,-0.613786,-0.251014])
R0=np.concatenate([R01,R02,R03])
```

```python
Ts,Rs=RK4(diffeq,R0,h=0.00001,maxt=7) #period is 6.234671
```

```python
def plot3(Ts,Rs,trailing=True,filename='test'):
    X1s=Rs[:,0]
    Y1s=Rs[:,1]
    X2s=Rs[:,4]
    Y2s=Rs[:,5]
    X3s=Rs[:,8]
    Y3s=Rs[:,9]
        
    fig,ax=plt.subplots()
    
    ax.set_xlabel('X coordinate')
    ax.set_ylabel('Y coordinate')
    ax.set_title('Evolution of the Three Stars')
    
    #set the limits of the plot to be the range of the points with a buffer zone
    allX=np.concatenate([X1s,X2s,X3s])
    allY=np.concatenate([Y1s,Y2s,Y3s])
    Xmax=allX.max()
    Xmin=allX.min()
    Ymax=allY.max()
    Ymin=allY.min()
    width=Xmax-Xmin
    height=Ymax-Ymin
    ax.set_xlim(Xmin-0.1*width,Xmax+0.1*width)
    ax.set_ylim(Ymin-0.1*height,Ymax+0.1*height)
    
    # want 1000 frames for the gif
    step = round(len(Ts)/1000)
    frame_indices = range(0, len(Ts), step)

    if trailing==True:
        trail_length=step*50
    
    #initialize empty line and scatter plot
    line1, = ax.plot([], [], color='lightpink', zorder=2)
    scatter1 = ax.scatter([], [], color='deeppink', zorder=3) #zorder makes sure the dot is on top of the line
    line2, = ax.plot([], [], color='lightgreen', zorder=2)
    scatter2 = ax.scatter([], [], color='green', zorder=3)
    line3, = ax.plot([], [], color='lightblue', zorder=2)
    scatter3 = ax.scatter([], [], color='blue', zorder=3)
    
    def animate(frame):
        index=frame*step
    
        line_step=100
        if trailing==False:
            line1.set_data(X1s[:index+1:line_step],Y1s[:index+1:line_step])
            line2.set_data(X2s[:index+1:line_step],Y2s[:index+1:line_step])
            line3.set_data(X3s[:index+1:line_step],Y3s[:index+1:line_step])
        elif trailing==True:
            startindex=index-trail_length
            if startindex<0:
                startindex=0
            line1.set_data(X1s[startindex:index+1:line_step],Y1s[startindex:index+1:line_step])
            line2.set_data(X2s[startindex:index+1:line_step],Y2s[startindex:index+1:line_step])
            line3.set_data(X3s[startindex:index+1:line_step],Y3s[startindex:index+1:line_step])
        
        scatter1.set_offsets([[X1s[index], Y1s[index]]])
        scatter2.set_offsets([[X2s[index], Y2s[index]]])
        scatter3.set_offsets([[X3s[index], Y3s[index]]])
        
        return line, scatter
    
    anim=animation.FuncAnimation(fig,animate,frames=len(frame_indices),
                                 interval=50,repeat=True,blit=True)
    
    anim.save(f'{filename}.gif',writer='pillow',fps=20)
```

```python
plot3(Ts,Rs,filename='test2')
```
