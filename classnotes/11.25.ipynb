New topic: numerical gravity
- nice!
- weakest of the four forces but functions at the furthest scales, so in astrophysical problems it is the dominant force
- four regimes that are handled differently:
    - planetary systems
    - star clusters
    - galaxies and cosmology
    - merging black holes
- first step for any problem is to figure out what you have to include and what you don't, like the smaller masses or general relativity
- that adds a second level of approximation error, since the math equations we are estimating numerically are an approximation of what is actually physically happening

Planetary systems:
- with just one star and ignoring the masses of the planets, it's an exactly solved problem - Kepler's Laws
- but the planets do have influence on each other - Neptune was discovered because of how it perturbed Uranus's orbit
- can only accurately model our solar system out to 100M years since it isn't completely stable
- key issue: inner planets orbit much more quickly than outer planets, so they need different time steps to accurately model them but also not waste computational time on the outer planets 
- can't just consider the planets separately though because their impact on each other at each time needs to stay lined up
- see slides for example simulation of planet behaving chaotically in stable two body star system

Chaos:
- systems where small changes in initial conditions lead to exponentially divergent outcomes
- many systems are chaotic
- need to be careful when interpreting the simulation results
- ex: simulate the evolution of the system many different times with slightly different initial positions within the known uncertainties, which can be used to learn how chaotic the system is and not the actual future positions of the system

Extended bodies:
- considering the stars and planets to just be points instead of spheres is also an approximation, but it only matters when things are getting close to each other and they can get ripped apart
- also if you want to model something as high-res as tidal forces between the earth and the moon
- to consider this, also need to add a model of hydrodynamics/geophysics 

Resonances:
- tend to occur in these long term repeating systems
- very hard to model though because their effects are very small but they repeat over and over again so the effects eventually add up, but so does the error
- often the resonance occurs in a small part of the phase space so that part has to be densely sampled 

Cosmology:
- the largest scale
- need to have particles to represent large amounts of material like millions of stars
- so the interaction between any two particles is meaningless, but the overall structure can still be accurate
- this is assuming that the stuff happening on the particle scale doesn't actually impact the larger scale
- first problem with N-boday: this scales as N^2 so that limits us to 100,000 particles
- can get around this by averaging out the affect of all the particles far away instead calculating the interaction between all the particles directly
- could do this with a tree of particles based on their spatial locations (like what I did for MSE) or do a grid and calculate how many particles are in each grid square
- need to be careful that the numerical implementation is not having an effect on the results
- don't need to use general relativity but do need to consider expansion of the universe over time

Relaxation time:
- estimate of the time it would take for a star to have its substantially changed by 2-body interactions (ie, basically colliding with another object)
- formula on slides
- divides gravitational systems into collisional and collisionless systems
- most astrophysical systems are collisionless - only dense stellar clusters have relaxation times that can be as short as 100Myr

Gravitational softening:
- if our system is collisonless, then 2 body encounters in our simulation are not physical and we want to remove them, since they'd show up because of the giant estimated particle size that actually represents many stars
- change gravitational force to be proportional to (r+epsilon)^-2 instead of r^-2 where epsilon is called the smoothing length
- that way we don't get the very large accelerations that would occur when particles are very close, because the real particles represented by the large particles are not getting very close
- if the system is collisional, then we need to resolve the encounters which are computationally very expensive because we need to use very short time steps
- problems like galaxy collisions or individual galaxies are collisionless 

Star clusters:
- actually a very difficult problem because these systems are not collisionless because of the density of the stars
- need very high resolution to make sure that the relevant 2-body interactions are modeled correctly
- need to consider more processes like supernovas and mass shedding of stars 

Black hole mergers:
- very relevant now that we can detect the gravitational waves produced by them
- easier than you might think because black holes have three properties: mass, spin, ?
- actually much more complicated if you have a neutron star - black hole merger because the internal structure of the neutron star actually matters a lot
- the emitted waves propagate infinitely so they are not affected by error that grows as you get further from the black holes
