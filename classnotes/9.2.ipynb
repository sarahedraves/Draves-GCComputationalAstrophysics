```python
import struct
import numpy as np

def write_mock_tipsy(filename, nstar=100):
    """
    Create a minimal TIPSY binary file with only star particles.
    Particles are randomly distributed in a Plummer-like sphere.

    Parameters
    ----------
    filename : str
        Output file path.
    nstar : int
        Number of star particles.
    """
    # TIPSY header: time (d), nbodies (i), ndim (i), nsph (i), ndark (i), nstar (i)
    time = 0.0
    ndim = 3
    nsph = 0
    ndark = 0
    nbodies = nstar
    header_fmt = 'diiiii'
    header_data = struct.pack(header_fmt, time, nbodies, ndim, nsph, ndark, nstar)

    # Generate random particle positions and velocities (simple Plummer sphere)
    np.random.seed(42)
    positions = np.random.normal(scale=1.0, size=(nstar, 3))
    velocities = np.random.normal(scale=0.1, size=(nstar, 3))
    masses = np.ones(nstar) * 1.0e6  # arbitrary mass per particle
    tform = np.zeros(nstar)
    eps = np.zeros(nstar)
    phi = np.zeros(nstar)
    metals = np.zeros(nstar)
    ids = np.arange(1, nstar + 1)
    ltime = np.zeros(nstar)
    rtime = np.zeros(nstar)

    # Star particle format:
    # pos(3d), vel(3d), mass(d), tform(d), eps(d), phi(d), metals(d), id(i), ltime(d), rtime(d)
    star_fmt = '6d d d d d d i d d'
    star_size = struct.calcsize(star_fmt)

    with open(filename, 'wb') as f:
        f.write(header_data)
        for i in range(nstar):
            data_tuple = (
                positions[i, 0], positions[i, 1], positions[i, 2],
                velocities[i, 0], velocities[i, 1], velocities[i, 2],
                masses[i], tform[i], eps[i], phi[i], metals[i],
                int(ids[i]), ltime[i], rtime[i]
            )
            f.write(struct.pack(star_fmt, *data_tuple))

    print(f"Mock TIPSY file '{filename}' written with {nstar} star particles.")
```

```python
def read_tipsy(filename):
    """
    Reads a TIPSY binary file and returns particle data as a dictionary.
    Assumes double precision (standard TIPSY format).
    """
    with open(filename, 'rb') as f:
        # Read header
        header_fmt = 'diiiii'  # time (double), nbodies, ndim, nsph, ndark, nstar (ints)
        header_size = struct.calcsize(header_fmt)
        header_data = f.read(header_size)
        time, nbodies, ndim, nsph, ndark, nstar = struct.unpack(header_fmt, header_data)

        # Skip gas (SPH) particles
        sph_size = nsph * (4*8 + 7*4 + 3*4 + 4 + 4 + 8)  # not reading details
        f.seek(sph_size, 1)

        # Read dark matter particles
        dark_fmt = '6d i d d d'  # pos(3d), vel(3d), mass(d), eps(d), phi(d), integer (id)
        dark_size = struct.calcsize(dark_fmt)
        f.seek(ndark * dark_size, 1)

        # Read star particles
        star_fmt = '6d d d d d d i d d'  # pos, vel, mass, tform, eps, phi, metals, id, ltime, rtime
        star_size = struct.calcsize(star_fmt)
        stars = []
        for _ in range(nstar):
            data = struct.unpack(star_fmt, f.read(star_size))
            pos = np.array(data[0:3])
            vel = np.array(data[3:6])
            mass = data[6]
            stars.append((pos, mass))

    positions = np.array([p for p, m in stars])
    masses = np.array([m for p, m in stars])
    return positions, masses


def stellar_mass_radial_surface_profile(filename, nbins=50, rmax=None):
    """
    Computes the stellar mass surface density profile from a TIPSY file.
    
    Parameters
    ----------
    filename : str
        Path to TIPSY binary snapshot.
    nbins : int
        Number of radial bins.
    rmax : float or None
        Maximum radius for profile. If None, determined from data.
    
    Returns
    -------
    r_mid : np.ndarray
        Midpoint radius of bins.
    sigma : np.ndarray
        Stellar surface density [mass per unit area] in each annulus.
    """
    pos, mass = read_tipsy(filename)
    
    # Cylindrical radius in XY-plane
    r = np.sqrt(pos[:,0]**2 + pos[:,1]**2)
    
    if rmax is None:
        rmax = r.max()

    # Bin edges and areas
    bins = np.linspace(0, rmax, nbins+1)
    areas = np.pi*(bins[1:]**2 - bins[:-1]**2)
    
    # Mass per radial annulus
    mass_profile, _ = np.histogram(r, bins=bins, weights=mass)
    
    # Surface density
    sigma = mass_profile / areas
    r_mid = 0.5 * (bins[1:] + bins[:-1])
    
    return r_mid, sigma
```

```python
write_mock_tipsy("mock_galaxy.bin", nstar=500)

# Then compute profile
r, sigma = stellar_mass_radial_surface_profile("mock_galaxy.bin", nbins=40)

import matplotlib.pyplot as plt
plt.plot(r, sigma)
plt.xlabel("Radius")
plt.ylabel("Surface Density")
plt.show()
```
