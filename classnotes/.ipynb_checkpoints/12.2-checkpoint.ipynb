Overview of Numerical Simulations:
- hydrodynamic simulations are important in astrophysics because of modeling gas
- this is still a simplification because we are ignoring magnetic effects
- people often misuse the term simulation to mean modeling
- the key thing that makes something a numerical simulation is that one does not know the future state of the system on which the rules will be applied
- often the rules are ODEs or PDEs but it could be something else

What is an element:
- like what is a single point in your simulation
- with something like orbiting planets, each planet is a point, but if you want to consider tidal forces, then the elements would have to be chunks of the planet
- in fluid mechanics, need to break up the fluid into different volumes where the properties of the fluids have different values
- could make a grid of space and then have the fluid flowing through the grid - Eulerian approach
- alternatively, could focus on a mass element of the fluid and track where it moves - Lagrangian approach

Equations of fluid dynamics:
- based on the three main conservation laws: mass, momentum, energy
- all the equations can be expressed in either the Eulerian or Lagrangian way
- these equations are very long and complicated and on the slides
- there are different simplifications possible like ignoring viscosity or having the fluid be incompressible depending on the situation
- also need an equation of state, which is the relationship between density and pressure in that specific fluid - the ideal gas law is an example

Discretization:
- finite-difference method involves creating a grid and only solving for the values of our function at the grid points
- finite-volume involves creating a grid but the values of the grid points are the volume averaged value of the function in that cell, which is better for situations with conservation laws
- calculating derivatives is different with the finite-volume approach if you want more than 2nd order accuracy
- need to fit a smooth function across multiple cells to do the derivatives
- for the derivatives, thinking about flux through the sides of the cells, which is why it's inherently conservative
- need to have ghost cells at the end of the real cells that have flux flow with the rest of the cells but the properties are set by the boundary conditions - periodic ones are common
- can make the grids super complicated:
    - chop up the grid for parallelization
    - adapting resolution of grid based on importance
    - other shaped grids - maybe cylindrical or spherical coordinates
    - grid can move in space - kind of hybrid with Lagrangian approach
- approaches should be chosen based on the problem
- discretizing the time derivative is a difficult problem also
    - time step can't be too large because if a fluid flows across more than one cell in a time step the errors will get really large
    - use the Courant condition to figure out how large the time step needs to be relative to the grid size
- smooth particle hydrodynamics: discretize the fluid by mass instead of volume
    - the fluid properties are associated with the fixed mass particles
    - the fluid properties are smoothed over nearby particles, with N nearest neighbors and a Gaussian kernel or a different kernel
    - Lagrangian method
    - each step, move the particles and then recalculate properties
    - has issues with shock capturing and creating artificial surface tension
    - computationally efficient because the calculations are done where the mass is - basically a cheaper way of getting high resolution but not as accurate as adaptive resolution finite volume
- unlike some other problems we've looked at, like n-body, where the different approaches converge, the results here will really depend on the approach chosen because the problems are so complex, so the choices are super important
