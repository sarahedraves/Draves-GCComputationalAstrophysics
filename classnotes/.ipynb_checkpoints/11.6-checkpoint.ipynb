Initial value problems:
- have time dimension and at least one spatial dimension
- have value of the quantity all along the spatial dimension at a specific time
- want the value of the quantity at a later time
- can't solve like the boundary value problems (with instead a time-space 2D grid), because we don't know the end point in the time dimension to use as a constraint
- need an iterative process like with the ODEs
- forward-time centered space method:
    - divide the spatial dimensions into a grid of points
    - then each point can be thought of as its own time-domain ODE
    - normally just use Euler's method because the ODE is already an approximation now that the spatial grid has been introduced
    - the centered space part means that the partial derivative with respect to the spatial derivative depends on all the adjacent points on the grid
    - the forward time part just means that the current step depends on the previous one

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#constants
L = 0.01 # Thickness of steel in meters
D = 4.25e-6 # Thermal diffusivity in meters squared per second
N = 100 # Number of divisions in grid
a = L/N # Grid spacing (because there are 100 divisions across the 1cm thickness)
h = 1e-4 # Time-step
epsilon = h/1000 #just the tolerance for the time comparisons
```

```python
#initial conditions for temperature
Tlo = 0.0 # Low temperature in Celcius
Tmid = 20.0 # Intermediate temperature in Celcius
Thi = 50.0 # Hi temperature in Celcius
T = np.empty(N+1,float)
T[0] = Thi #anchor end points to the temperature reservoirs
T[N] = Tlo
T[1:N] = Tmid #all other points in the grid are the initial temperature of the steel
Tp = T.copy() #T prime array for filling with new T values and swapping
```

```python
#times for plotting in seconds
t1 = 0.01
t2 = 0.1
t3 = 0.4
t4 = 1.0
t5 = 10.0
tend = t5 + epsilon #will have a while loop exit when the end time is reached, right after t5 has been calculated
```

```python
# Main loop
t = 0.0 #initial time
c = h*D/(a*a)
while t<tend:
    # Calculate the new values of T
    #note that the previous time step Ts are not stored
    #which is why the plotting is done in the while loop below
    for i in range(1,N): #the end points are unchanged due to contact with temp reservoirs
        # and b/c the centered difference would not allow them to be calculated due to index error
        Tp[i] = T[i] + c*(T[i+1]+T[i-1]-2*T[i]) #see slide 22 for source of this equation
        #basically the step/slope is determined by the values of the points around this one in this time step
    
    T,Tp = Tp,T #switch the temp arrays for efficiency of memory usage
    t += h #take time step
    
    # Make plots at the given times
    if abs(t-t1)<epsilon:
        plt.plot(T)
    if abs(t-t2)<epsilon:
        plt.plot(T)
    if abs(t-t3)<epsilon:
        plt.plot(T)
    if abs(t-t4)<epsilon:
        plt.plot(T)
    if abs(t-t5)<epsilon:
        plt.plot(T)
```

```python
#exercise 9.4
#constants
tau=365 #days
A=10.0 #degrees C
B=12.0 #degrees C
T20=11.0 #degrees C - temp at depth of 20m year round
D=0.1 #thermal diffusivity of earth's crust in meters squared per day
```

```python
def T0(dayofyear): #T0 as in z=0, not t=0
    return A+B*np.sin(2*np.pi*dayofyear/tau)
```

```python
#create starting temp array with t=0
N=100 #number of grid divisions
T=np.empty(N+1,float)
T[0]=T0(0)
T[N]=T20
T[1:N]=10.0 #starting temp. T0 as in t=0, not z=0.
Tp=T.copy() #T prime array for filling with new T values and swapping
```

```python
L = 20 #because depth is up to 20m
a = L/N # Grid spacing (because there are 100 divisions across the 1cm thickness)
Z=np.arange(0,L+a,a) #this array is just for plotting, but it's the depth of each grid point
```

```python
h = 1e-2 #time step in days
epsilon = h/1000 #just the tolerance for the time comparisons
stepsperday=1/h
tend=10*tau+epsilon #end time of 10 years
```

```python
tplot=9*tau+epsilon #start plotting after nine years have been passed
#only plot on the days of the year below
d1=0
d2=91
d3=182
d4=274
d5=365
```

```python
# Main loop
t = 0.0 #initial time
c = h*D/(a*a)
while t<tend:
    dayofyear=t%tau 
    Tp[0]=T0(dayofyear) 
    
    # Calculate the new values of T
    #note that the previous time step Ts are not stored
    #which is why the plotting is done in the while loop below
    for i in range(1,N): #the end points are unchanged due to contact with temp reservoirs
        # and b/c the centered difference would not allow them to be calculated due to index error
        Tp[i] = T[i] + c*(T[i+1]+T[i-1]-2*T[i]) #see slide 22 for source of this equation
        #basically the step/slope is determined by the values of the points around this one in this time step
    
    T,Tp = Tp,T #switch the temp arrays for efficiency of memory usage
    t += h #take time step

    if t>tplot: #only plot after 9 years have passed
        # Make plots on the given days
        if abs(dayofyear-d1)<epsilon:
            plt.plot(Z,T)
        if abs(dayofyear-d2)<epsilon:
            plt.plot(Z,T)
        if abs(dayofyear-d3)<epsilon:
            plt.plot(Z,T)
        if abs(dayofyear-d4)<epsilon:
            plt.plot(Z,T)
        if abs(dayofyear-d5)<epsilon:
            plt.plot(Z,T)
```
