Technological applications:
- JPEG - does a 2D transform of the image and stores the coefficients
    - also drops some of the smaller coefficients
    - makes the file smaller for only a small information loss
    - MPEG does the same thing for movies
- MP3 is similar but DFT is done in time instead of space
    - also chooses information based on which frequencies are audible to humans
    - also specific things like loud low frequency sounds means humans can't hear high frequency sounds as well so it can keep fewer of those
    - modern ones are encoded to be at the limit of what humans can actually hear and it's not dropping more data than that

Fast fourier transforms:
- DFT requires 1/2N^2 calculations, so if N is about 45,000 it will require a billion calculations and that's only 1 second of music
- FFT is much better
- basically, any fourier transform can be written as a sum of two fourier transforms with half as many points
- then split those fourier transforms into two with half as many points
- repeat until each fourier transform has only one point - then it's trivial to perform
- so scales at Nlog2(N) which is as could as it gets
- the code is much more complicated than the basic DFT so we won't bother coding it
- Numpy has implementations for this - can use this for the homework
- if the points are real, can use rfft() and  irfft() - do not mix and match real vs complex versions of the functions because then the number of cofficients will be off

- Power spectrum is the modulus squared of the coefficients of the a Fourier transform on a infinitely precise function
- Periodogram is the same thing but on discrete data so it's really an estimate of the power spectrum
- For observational data, can only have a periodogram, but people often interchange the terms

Random processes:
- Computers cannot actually pick a random number (but humans can't really either)
- Observing radioactive decay is probably the best way to get true random numbers
- Computers make pseudorandom numbers where they are generated by a deterministic formula
- Basic version:
    - x'=(ax+c) mod m
    - a,c,m are integer constants are x is an interger variable
    - take an x and then plug it into the equation to get x' and the plug it back into the equation
    - doing this repeatedly will give numbers that appear to be randomly generated if a,c,m are specific integers
        - ex: if c and m are both even, then the results will be either only even or only odd depending on initial x
    - produces values in the frange from 0 to m-1
    - x is called the random seed which is useful when testing code
    - this method is called a linear congruential random number generator
    - these numbers aren't random enough for many physics uses and also for gambling games
- Often when numbers aren't random enough, it's because the order that terms are generated in leaves a time-series correleration, even though the final set of numbers is randomly distributed
- Popular physics choice: Mersenne twister, a generalized feedback shift-register generator
- Coming up with better random number generators is a whole field
- In Python, use the numpy random subpackage instead of just the random package

Cryptography:
- often relies on random numbers
- the random numbers are used to switch out each letter by a different shift in the alphabet
- can only decode the message if you know the order of the random numbers used
- practical implementation is to have a random number generator make the random numbers, and then have the seed as a key
