NB: getting the code to actually run is just the first step! Then you have to make sure you actually believe that the results are meaningful.

Cross-validation:
- great candidate for parallelization if the number of splits in the data is the same as the number of cores requested
- can average the results from the cross sections
- can get slow quickly though
- can't improve past a certain point though - once the train and test scores start to converge on the learning curve 

```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold, cross_validate, cross_val_predict, learning_curve
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('planets')
```

```python
df=df.drop(['method','number'],axis=1)
df.dropna(inplace=True)
```

```python
df['logMass']=np.log10(df['mass'])
df['logDistance']=np.log10(df['distance'])
```

```python
df
```

```python
df.corr()
```

```python
X=df[['orbital_period','year','logDistance']].values
```

```python
y=df['logMass'].values.reshape(-1,1)
```

```python
kf = KFold(shuffle = True, n_splits = 5, random_state=5)
```

```python
model=linear_model.LinearRegression()
```

```python
mass_scores = cross_validate(model, X, y, cv = kf, return_train_score = True)
```

```python
print('Test scores:','{:.3f}'.format(mass_scores['test_score'].mean()), '{:.3f}'.format(mass_scores['test_score'].std()))
print('Train scores:','{:.3f}'.format(mass_scores['train_score'].mean()), '{:.3f}'.format(mass_scores['train_score'].std()))
```

```python
mass_predictions=cross_val_predict(model, X, y, cv = kf)
```

```python
plt.scatter(y, mass_predictions)
plt.plot(y,y,color='black')
plt.xlabel('Actual Log Mass')
plt.ylabel('Predicted Log Mass')
plt.show()
```

```python
train_sizes, train_scores, test_scores=learning_curve(model, X, y, cv=kf)
```

```python
train=train_scores.mean(axis=1)
train_std=train_scores.std(axis=1)
test=test_scores.mean(axis=1)
test_std=test_scores.std(axis=1)
```

```python
fig,ax=plt.subplots()
ax.plot(train_sizes,train)
ax.plot(train_sizes,test)
ax.fill_between(train_sizes,train+train_std,train-train_std,alpha=0.3)
ax.fill_between(train_sizes,test+test_std,test-test_std,alpha=0.3)
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfmodel=RandomForestRegressor()
```

```python
y=y.reshape(y.shape[0])
```

```python
mass_scores = cross_validate(rfmodel, X, y, cv = kf, return_train_score = True)
```

```python
print('Test scores:','{:.3f}'.format(mass_scores['test_score'].mean()), '{:.3f}'.format(mass_scores['test_score'].std()))
print('Train scores:','{:.3f}'.format(mass_scores['train_score'].mean()), '{:.3f}'.format(mass_scores['train_score'].std()))
```

```python
mass_predictions=cross_val_predict(rfmodel, X, y, cv = kf)
```

```python
plt.scatter(y, mass_predictions)
plt.plot(y,y,color='black')
plt.xlabel('Actual Log Mass')
plt.ylabel('Predicted Log Mass')
plt.show()
```
